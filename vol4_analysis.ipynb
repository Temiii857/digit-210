```python
import nltk
import matplotlib.pyplot as plt
from nltk import FreqDist
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

```python
with open("vol-4.txt", "r", encoding="utf-8") as file:
    vol4_text = file.read()

print(vol4_text[:300])
```

```python
tokens = word_tokenize(vol4_text)
words = [word.lower() for word in tokens if word.isalpha()]  # only keep real words, no punctuation

print(words[:20])
```

```python
def lexical_diversity(text):
    return len(set(text)) / len(text)

print("Lexical diversity of vol-4.txt:", lexical_diversity(words))
```

```python
fdist = FreqDist(words)
fdist.plot(20, title="Top 20 Words in vol-4.txt")
plt.show()
```
